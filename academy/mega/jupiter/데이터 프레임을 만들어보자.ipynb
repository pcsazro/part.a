# 행을 합하는 함수 : concat, append
* concat은 여러개를 합치는 경우
* append는 두개를 합치는 경우

```python
df2 = pd.read_csv('data/concat_1.csv')
df2
```

```python
df3 = pd.read_csv('data/concat_2.csv')
df3
```

```python
df4 = pd.read_csv('data/concat_3.csv')
df4
```

```python
row_concat = pd.concat([df2,df3,df4],ignore_index=True)
row_concat
```

```python
df2 = pd.read_csv('data/concat_1.csv')
df3 = pd.read_csv('data/concat_2.csv')
df4 = pd.read_csv('data/concat_3.csv')
```

```python
df2
```

```python
df3
```

```python
df4
```

```python
row_concat = pd.concat([df2,df3,df4],ignore_index=True)
```

```python
row_concat
```

```python
row_append = df2.append(df3)
```

```python
row_append
```

```python
row_append2 = df3.append(df4, ignore_index=True)
```

* append 함수를 이용하여 row를 추가하는 경우
     추가되는 대상이 dataframe이어야 한다.

```python
row_append2
```

* 한 row만 넣고 싶은 경우, dictionary로 만들어 추가한다.

```python
new_dic = {'A':'new1','B':'new2','C':'new3','D':'new4'}
```

```python
df5 = df2.append(new_dic,ignore_index=True)
df5
```

* outer join

```python
result1 = pd.concat([df2,df3,df4])
result1
```

* inner join

```python
result2 = pd.concat([df2,df3,df4], join='inner')
result2
```

```python
result3 = pd.concat([df2,df4], join='inner', ignore_index=True)
result3
```

```python
result4 = pd.concat([df3,df4], join='inner', ignore_index=True)
result4
```

# 열을 합치는 함수 : merge
* 하나의 동일한 컬럼이 있어야한다.
* left_join된다.

```python
result5 = df2.merge(df4, on='A')
```
