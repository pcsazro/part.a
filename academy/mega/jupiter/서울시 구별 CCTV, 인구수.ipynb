```python
fp1 = np.polyfit(data_result['총 인구 합계'], data_result['소계'], 1)
```

```python
fp1
```

array([      기울기,        절편     ])

```python
fx = np.linspace(100000, 700000, 100)
fy = np.poly1d(fp1)
```

```python
plt.figure(figsize=(10,10))

plt.scatter( data_result['총 인구 합계'], data_result['소계'], s=50)
plt.xlabel('총 인구')
plt.ylabel('CCTV')
plt.grid()

plt.plot(fx, fy(fx), ls='dashed', lw=3, color='y')

plt.show()
```

각각의 점에 글자 넣기

```python
plt.figure(figsize=(10,10))

plt.scatter( data_result['총 인구 합계'], data_result['소계'], s=50)
plt.xlabel('총 인구')
plt.ylabel('CCTV')
plt.grid()

plt.plot(fx, fy(fx), ls='dashed', lw=3, color='y')

for n in range(25):
    plt.text(data_result['총 인구 합계'][n]*1,
             data_result['소계'][n]*1.02,
             data_result.index[n], fontsize=13)

plt.show()
```
